### Datos Faltantes

```python
import pandas as pd
import numpy as np

# Crear un DataFrame de ejemplo
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [np.nan, 2, 3, np.nan, 5],
    'C': [1, 2, 3, 4, 5]
}

df = pd.DataFrame(data)
print("DataFrame original:")
df

```

```python
df_dropped_rows = df.dropna()
print("\nDataFrame después de eliminar \
      filas con valores faltantes:")
df_dropped_rows
```

```python
df_dropped_cols = df.dropna(axis=1)
print("\nDataFrame después de eliminar columnas con valores faltantes:")
df_dropped_cols
```

```python
df['A'] = df['A'].fillna(df['A'].mean())  
df['B'] = df['B'].fillna(df['B'].median())  
df['C'] = df['C'].fillna(df['C'].mode().iloc[0])  

df
```

```python
import pandas as pd
import numpy as np

# Crear un DataFrame de ejemplo
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [np.nan, 2, 3, np.nan, 5],
    'C': [1, 2, 3, 4, 5]
}

df = pd.DataFrame(data)
print("DataFrame original:")
print(df)

# 1. Eliminar Filas con Valores Faltantes
df_dropped_rows = df.dropna()
print("\nDataFrame después de eliminar filas con valores faltantes:")
print(df_dropped_rows)

# 2. Eliminar Columnas con Valores Faltantes
df_dropped_cols = df.dropna(axis=1)
print("\nDataFrame después de eliminar columnas con valores faltantes:")
print(df_dropped_cols)

# 3. Rellenar con la Media de la Columna
df_filled_mean = df.fillna(df.mean())
print("\nDataFrame después de rellenar con la media de cada columna:")
print(df_filled_mean)

# 4. Rellenar con la Mediana de la Columna
df_filled_median = df.fillna(df.median())
print("\nDataFrame después de rellenar con la mediana de cada columna:")
print(df_filled_median)

# 5. Rellenar con un Valor Constante
df_filled_constant = df.fillna(0)
print("\nDataFrame después de rellenar con un valor constante (0):")
print(df_filled_constant)

# 6. Interpolación
df_interpolated = df.interpolate()
print("\nDataFrame después de la interpolación:")
print(df_interpolated)
```
